# Read file using path

## mode
========= ===============================================================
Character Meaning
--------- ---------------------------------------------------------------
'r'       open for reading (default)
'w'       open for writing, truncating the file first
'x'       create a new file and open it for writing
'a'       open for writing, appending to the end of the file if it exists
'b'       binary mode
't'       text mode (default)
'+'       open a disk file for updating (reading and writing)
========= ===============================================================

```python
# windows path - C:\\MyPC\\personal\\github\\python\\basics\\file2.txt
# Linux/Mac  - /Users/UserName/personal/github/python/basics/file2.txt
# with below syntax we don't need to call close() manually
with open("C:\\MyPC\\personal\\github\\python\\basics\\file2.txt", mode='r') as file2:
    contents = file2.read()
    print(contents)
```

```python
with open("C:\\MyPC\\personal\\github\\python\\basics\\file2.txt", mode='a') as file:
    file.write('\nthis is fourth line')
```

```python
with open("C:\\MyPC\\personal\\github\\python\\basics\\file2.txt", mode='r') as file:
    print(file.read())
```

```python
with open("randomFile.txt", mode='w') as f:
    f.write("This is random file")
```

```python
with open("randomFile.txt", mode='r') as f:
    print(f.read())
```

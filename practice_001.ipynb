## File Mode Characters in Python open()

| Character | Meaning                                                        | Example Usage                          |
|-----------|----------------------------------------------------------------|----------------------------------------|
| `r`       | Open for reading (default)                                     | `open("file.txt", "r")`                |
| `w`       | Open for writing, truncating the file first                    | `open("file.txt", "w")`                |
| `x`       | Create a new file and open it for writing                      | `open("newfile.txt", "x")`             |
| `a`       | Open for writing, appending to the end of the file if it exists| `open("file.txt", "a")`                |
| `b`       | Binary mode                                                    | `open("image.png", "rb")`              |
| `t`       | Text mode (default)                                            | `open("file.txt", "rt")`               |
| `+`       | Open a disk file for updating (reading and writing)            | `open("file.txt", "r+")` or `open("file.txt", "w+")` |

```python
# To create new file. Works only from notebook
# %%writefile file-1.txt
# This is first line
# This is second line
# This is third line

# To create new file
# If \n is not given then writes all in single line 
with open("file-1.txt", "w") as file:
    file.write("This is first line")
    file.write("This is second line")
    file.write("This is third line")

# Add each line in new line
with open("file-2.txt", "w") as f:
    print("This is first line", file=f)
    print("This is second line", file=f)
    print("This is third line", file=f)

# Add OS specific line separator. similar to Java Files.newLine
import os
with open("file-3.txt", "w") as f:
    f.write("This is first line" + os.linesep)
    f.write("This is second line" + os.linesep)
    f.write("This is third line" + os.linesep)
```

# palindrome

```python
word1 = 'racecar'
word2 = 'kayak'
word3 = 'step on no pets'
word4 = 'hello'

def is_palindrome(word):
    word = word.replace(' ', '')
    reverse_word = word[::-1]
    return word == reverse_word

list1 = [word1, word2, word3, word4]
for word in list1:
    result = is_palindrome(word)
    if result:
        print(f"'{word}' is palindrome")
    else:
        print(f"'{word}' is not palindrome")
```

```python
# pass input as tuples to class constructor and use them in method
class Line():

    # coordinates1, coordinates2 are tuples
    def __init__(self, coordinates1, coordinates2):
        print(f"type(coordinates1)={type(coordinates1)}, type(coordinates2)={type(coordinates2)}")
        self.coordinates1 = coordinates1
        self.coordinates2 = coordinates2

    def calculate_distance(self):
        x1,y1 = self.coordinates1
        x2,y2 = self.coordinates2

        print(f"x1={x1}, y1={y1}, x2={x2}, y2={y2}")
        return ((x2-x1)**2 + (y2-y1) ** 2) ** 0.5

    def slope(self):
        x1, y1 = self.coordinates1
        x2, y2 = self.coordinates2
        print(f"x1={x1}, y1={y1}, x2={x2}, y2={y2}")
        return (y2 - y1) / (x2 - x1)

line1 = Line((1, 2), (3, 4))
distance = line1.calculate_distance()
print(f"distance={distance}")
slope = line1.slope()
print(f"slope={slope}")
```

# Account deposit withdraw

```python
class Account():

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def __str__(self):
        return f"Account owner={self.owner}, Account balance={self.balance}"

    def deposit(self, amount):
        self.balance = self.balance + amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Funds unavailable")
        else:
            print("Withdraw accepted")
            self.balance = self.balance - amount

account = Account("a", 100)
print(account)

print(f"owner={account.owner}")
print(f"balance={account.balance}")

account.deposit(10)
print(f"balance={account.balance}")
account.deposit(10)
print(f"balance={account.balance}")
account.deposit(10)
print(f"balance={account.balance}")
account.deposit(10)
print(f"balance={account.balance}")
account.deposit(10)
print(f"balance={account.balance}")

account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(150)
print(f"balance={account.balance}")
```

```python
class Student():

    def __init__(self, fName, lName):
        self.firstName = fName
        self.lastName = lName

    def printName(self):
        self.name = self.firstName + ', ' + self.lastName
        print(f"name={self.name}")

student1 = Student('a', 'b')
print(f"{type(student1)}")
print(f"firstName={student1.firstName}, lastName={student1.lastName}")
student1.printName()

student2 = Student(fName='c', lName='d')
print(f"{type(student2)}")
print(f"firstName={student2.firstName}, lastName={student2.lastName}")
student2.printName()
```

```python
class Animal():
    def __init__(self):
        print("Animal created")

    def who_am_i(self):
        print("I am animal")

    def eat(self):
        print("I am eating")

animal1 = Animal()
animal1.who_am_i()
```

```python
# inherit from Animal
class Dog(Animal):
    
    def __init__(self):
        Animal.__init__(self)
        print("Dog created")

    def who_am_i(self):
        print("I am dog")

dog1 = Dog()

dog1.who_am_i()
dog1.eat()
```

```python
# abstract method
# raise exception if abstract method called

class Animal():

    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement this method")

# animal = Animal("animal")
# animal.speak() # this call throws exception

class Dog(Animal):

    def speak(self):
        print(self.name + " says woof")

class Cat(Animal):

    def speak(self):
        print(self.name + " says meow")

dog1 = Dog("dog1")
dog1.speak()

cat1 = Cat("cat1")
cat1.speak()
```

```python
#  Magic/Dunder methods
class Book():

    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    # similar to equals() in Java
    def __str__(self):
        return f"{self.title} by {self.author}"

    def __len__(self):
        return self.pages

    def __del__(self):
        print("Book is deleted")

book1 = Book('book1', 'author1', 100)
print(book1)
print(str(book1))
del(book1)
```

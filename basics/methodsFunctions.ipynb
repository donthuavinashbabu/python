# args and kwargs
* `*args` - arguments - to pass variable number of arguments. Goes to method as tuple
* `**kwargs` - keyword arguments - to pass variable number of arguments. To pass key value pairs. Goes to method as dictionary

```python
def function1(*args):
    print(args)

function1(40,60)
function1(10,20,30,40,60)
```

```python
def function2(**kwargs):
    print(kwargs)
    for key,value in kwargs.items():
        print(f"key={key}, value={value}")

function2(a=1)
function2(a=1, b=2, c=3)
```

```python
def function3(*args, **kwargs):
    print(args)
    print(kwargs)

function3(10,20,30,a=1,b=2,c=3)
```

# String enumerate upper lower function example
Define a function called myfunc that takes in a string, and returns a matching string where every even letter is uppercase, and every odd letter is lowercase. Assume that the incoming string only contains letters, and don't worry about numbers, spaces or punctuation. The output string can start with either an uppercase or lowercase letter, so long as letters alternate throughout the string.

Remember, don't run the function, simply provide the definition.

To give an idea what the function would look like when tested:

myfunc('Anthropomorphism')
Output: 'aNtHrOpOmOrPhIsM'
Added note: this exercise requires that the function return a string. Print statements will not work here.

```python
def myfunc(str1):
    result = ''
    for i, letter in enumerate(str1):
        if i % 2 == 0:
            result = result + letter.upper()
        else:
            result = result + letter.lower()
    return result

result1 = myfunc('Anthropomorphism')
print(result1)
```

## print each word in lits is even length or odd length

```python

def even_odd(word):
    if len(word) % 2 == 0:
        return "Even"
    else:
        return "Odd"

words = ['a', 'bb', 'ccc', 'dddd']
list1 = list(map(even_odd, words))
print(f"list1={list1}")
```

## filter list

```python
def checkEven(num):
    return num % 2 == 0

nums = [1,2,3,4,5,6]
list1 = list(filter(checkEven, nums))
print(f"list1:{list1}")

for num in filter(checkEven, nums):
    print(num)
```

## square with lambda

```python
def square1(num):
    result = num ** 2
    return result

# lambda version of above method
square2 = lambda num: num ** 2

print(f"square1(4)={square1(4)}")
print(f"square2(4)={square2(4)}")

nums = [1,2,3,4,5]
list1 = list(map(lambda num: num **2, nums))
print(f"list1={list1}")

nums2 = [1,2,3,4,5,6,7,8]
list2 = list(filter(lambda num: num%2 == 0, nums2))
print(f"list2={list2}")
```

# Dictionaries - key value pairs

```python
dict1 = {'a': '1', 'b': '2', 'c': ['3', '4'], 'd': {'e': '5', 'f': '6'}}
print(f"dict1={dict1}")

# get value
print(f"a value={dict1['a']}")
print(f"c value={dict1['c']}")
print(f"d value={dict1['d']}")
print(f"d value={dict1['d']['e']}")
```

```python
dict1 = {'a': 1, 'b': 2}
print(f"dict1={dict1}")
dict1['c']=3
print(f"dict1={dict1}")
dict1['b']=4
print(f"dict1={dict1}")
```

# Get all keys only, values only, pairings only of dictionaty

```python
dict1 = {'a': 1, 'b': 2}
print(f"dict1={dict1}")

keys = dict1.keys()
print(f"keys={keys}")

values = dict1.values()
print(f"values={values}")

items = dict1.items()
print(f"items={items}")
```

# zip

```python
list1 = ['a', 'b', 'c', 'd', 'e']
list2 = [1,2,3,4,5]
list3 = [100,200,300,400,500]
for item in zip(list1, list2, list3):
    print(f"{item}")
```

```python
list4 = list(zip(list1, list2, list3))
print(f"list4={list4}")
for a,b,c in list4:
    print(f"{a}={b}={c}")
```

# string to list

```python
str1 = "hello world"
list7 = list(str1)
print(f"list7={list7}")
```

```python
list8 = [i for i in str1]
print(f"list8={list8}")
```

```python
list9 = [num**2 for num in range(10)]
print(f"list9={list9}")
```

```python
list10 = [num for num in range(0, 11) if num % 2 == 0]
print(f"list10={list10}")
```

# multiplying list vales

```python
list11 = [x*y for x in [2,4,6] for y in [10,20,30]]
print(f"list11={list11}")
```

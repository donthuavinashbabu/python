```python
#  check word is upper or lower case

word1 = "hello"
word2 = "World"
word3 = "WORLD"

def isCapitalized(word):
    firstChar = word[0]
    return firstChar.isupper()
        
print("isupper")
print(f"word1={word1.isupper()}")
print(f"word2={word2.isupper()}")
print(f"word3={word3.isupper()}")

print("\nislower")
print(f"word1={word1.islower()}")
print(f"word2={word2.islower()}")
print(f"word3={word3.islower()}")

print("\nisCapitalized")
print(f"word1={isCapitalized(word1)}")
print(f"word2={isCapitalized(word2)}")
print(f"word3={isCapitalized(word3)}")
```

```python
# palindrome
word1 = 'racecar'
word2 = 'kayak'
word3 = 'step on no pets'
word4 = 'hello'

def is_palindrome(word):
    word = word.replace(' ', '')
    reverse_word = word[::-1]
    return word == reverse_word

list1 = [word1, word2, word3, word4]
for word in list1:
    result = is_palindrome(word)
    if result:
        print(f"'{word}' is palindrome")
    else:
        print(f"'{word}' is not palindrome")
```

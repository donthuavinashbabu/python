```python
# for loop
list1 = [1,2,3,4,5]
print(f"list1={list1}")
for i in list1:
    print(f"item={i}")
```

```python
# empty for loop
for i in list1:
    pass
```

```python
# don't want to use variable i, use _ instead
for _ in list1:
    print("hello")
```

```python
# print even numbers from 0 to 10
for i in range(0, 11):
    if i % 2 == 0:
        print(f"even number={i}")
    else:
        print(f"odd number={i}")
```

```python
# tuple
tup1 = (1,2,3)
print(f"tup1={tup1}")
for i in tup1:
    print(f"item={i}")
```

```python
# tuples in list
list2 = [(1,2), (3,4), (5,6)]
print(f"list2={list2}")
for i in list2:
    print(f"tup={i}")
    for j in i:
        print(f"item={j}")
```

```python
list2 = [(1,2), (3,4), (5,6)]
for (i,j) in list2:
    print(f"i={i}, j={j}")
```

```python
list2 = [(1,2), (3,4), (5,6)]
for i,j in list2:
    print(f"i={i}, j={j}")
```

```python
# dictionary
dict1 = {"a":1, "b":2, "c":3}
print(f"dict1={dict1}")
for key in dict1:
    print(f"key={key}, value={dict1[key]}")
```

```python
dict1 = {"a":1, "b":2, "c":3}
for item in dict1.items():
    print(f"item={item}")
```

```python
dict1 = {"a":1, "b":2, "c":3}
for key, value in dict1.items():
    print(f"key={key}, value={value}")
```

```python
# while loops
i = 0
while i < 5:
    print(f"i={i}")
    i = i + 1
else:
    print("i is no longer less than 5")
```

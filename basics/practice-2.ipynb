```python
# Keep adding numbers only to list
list1 = []

def ask_input():
    input1 = input("Enter any number: ")
    return input1

from IPython.display import clear_output
def play():
    # input1 = ask_input()
    continue_loop = True

    while continue_loop:
        input1 = ask_input()
        
        if input1 == '-1':
            break
        
        while input1.isdigit() == False:
            input1 = ask_input()
            print("Not a number. Try again")

        clear_output()
        list1.append(int(input1))
        print(f"{input1} added to list")

play()
print(f"list1={list1}")
```

```python
# pass input as tuples to class constructor and use them in method
class Line():

    # coordinates1, coordinates2 are tuples
    def __init__(self, coordinates1, coordinates2):
        print(f"type(coordinates1)={type(coordinates1)}, type(coordinates2)={type(coordinates2)}")
        self.coordinates1 = coordinates1
        self.coordinates2 = coordinates2

    def calculate_distance(self):
        x1,y1 = self.coordinates1
        x2,y2 = self.coordinates2

        print(f"x1={x1}, y1={y1}, x2={x2}, y2={y2}")
        return ((x2-x1)**2 + (y2-y1) ** 2) ** 0.5

    def slope(self):
        x1, y1 = self.coordinates1
        x2, y2 = self.coordinates2
        print(f"x1={x1}, y1={y1}, x2={x2}, y2={y2}")
        return (y2 - y1) / (x2 - x1)

line1 = Line((1, 2), (3, 4))
distance = line1.calculate_distance()
print(f"distance={distance}")
slope = line1.slope()
print(f"slope={slope}")
```

```python
class Account():

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def __str__(self):
        return f"Account owner={self.owner}, Account balance={self.balance}"

    def deposit(self, amount):
        self.balance = self.balance + amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Funds unavailable")
        else:
            print("Withdraw accepted")
            self.balance = self.balance - amount

account = Account("a", 100)
print(account)

print(f"owner={account.owner}")
print(f"balance={account.balance}")

account.deposit(10)
print(f"balance={account.balance}")
account.deposit(10)
print(f"balance={account.balance}")
account.deposit(10)
print(f"balance={account.balance}")
account.deposit(10)
print(f"balance={account.balance}")
account.deposit(10)
print(f"balance={account.balance}")

account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(10)
print(f"balance={account.balance}")
account.withdraw(150)
print(f"balance={account.balance}")
```

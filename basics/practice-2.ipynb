```python
# check input is integer or not
input1 = input("Enter number: ")
isDigit1 = input1.isdigit()
if isDigit1:
    print(f"{input1} is integer")
else:
    print(f"{input1} is not integer")
```

```python
# keep asking user to enter valid number
input1 = input("Enter any number: ")

while input1.isdigit() == False:
    input1 = input("Enter any number: ")

print(f"Number entered: {input1}")
```

```python
# Keep adding numbers only to list
list1 = []

def ask_input():
    input1 = input("Enter any number: ")
    return input1

from IPython.display import clear_output
def play():
    # input1 = ask_input()
    continue_loop = True

    while continue_loop:
        input1 = ask_input()
        
        if input1 == '-1':
            break
        
        while input1.isdigit() == False:
            input1 = ask_input()
            print("Not a number. Try again")

        clear_output()
        list1.append(int(input1))
        print(f"{input1} added to list")

play()
print(f"list1={list1}")
```
